```python
import numpy as np
import numpy.typing as npt
```

```python
class Deque:
    def __init__(self, capacidade: int) -> None:
        self.capacidade: int = capacidade
        self.inicio: int = -1
        self.final: int = 0
        self.numeros_elementos: int = 0
        self.valores: npt.ArrayLike = np.empty(self.capacidade, dtype=int)

    def __deque_cheio(self) -> bool:
        return (self.inicio == 0 and self.final == self.capacidade - 1) or (self.inicio >= self.final + 1)
    
    def __deque_vazio(self) -> bool:
        return self.inicio == -1
    
    def inserir_inicio(self, valor) -> None:
        if self.__deque_cheio():
            print("O deque está cheio")
            return
        if self.inicio == -1:
            self.inicio = 0
            self.final = 0
        elif self.inicio == 0:
            self.inicio = self.capacidade - 1
        else:
            self.inicio -= 1

        self.capacidade[self.inicio] = valor

    def inserir_final(self, valor) -> None:
        if self.__deque_cheio():
            print("O deque está cheio")
            return
        if self.inicio == -1:
            self.inicio = 0
            self.final = 0
        elif self.final == self.capacidade - 1:
            self.final = 0
        else:
            self.final += 1

        self.capacidade[self.inicio] = valor

    def excluir_inicio(self) -> None:
        if self.__deque_vazio:
            print("O deque está vazio")
            return
        
        if self.inicio == self.final:
            self.inicio -=1
            self.final -=1
        else:
            if self.inicio == self.capacidade - 1:
                self.inicio = 0
            else:
                self.inicio += 1
    
    def excluir_final(self) -> None:
        if self.__deque_vazio:
            print("O deque está vazio")
            return
        
        if self.inicio == self.final:
            self.inicio -=1
            self.final -=1
        elif self.inicio == 0:
            self.final = self.capacidade - 1
        else:
            self.final -= 1
    
    def get_inicio(self) -> int:
        if self.__deque_vazio:
            print("O deque está vazio")
            return
        return self.capacidade[self.inicio]
    
    def get_final(self) -> int:
        if self.__deque_vazio:
            print("O deque está vazio")
            return
        return self.capacidade[self.final]
```
